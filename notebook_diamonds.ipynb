## Step 1: Split the data
** Will split the data such that the cut is balanced among training and test**

````python
def split_data(data_df,category,nsplit,testsize,random=2301):
    trainsplit  =  StratifiedShuffleSplit ( n_splits = nsplit ,  test_size = testsize ,  random_state = random )
    
    """
    n_splits is for number of splits
    test_size is % size from the entire population(between 0 and 1) for the test data set
    random_state this is used to generate repeatable results next time we run this code again
    """
    for  train_index ,  test_index  in  trainsplit . split ( data_df ,  data_df[category]):
        strat_train_set  =  data_df.loc [train_index ]
        strat_test_set  =  data_df.loc[test_index ]
        
    """ 
    stratifying by classification variables
        
    """
    return strat_train_set,strat_test_set
````

````python
train_df,test_df =split_data(data_df=diam_df,category=['cut'],nsplit=1,testsize=0.2)
X_train=train_df[train_df.columns.difference(['cut', 'color', 'clarity','x','y','z','price'])].values

X_test=test_df[test_df.columns.difference(['cut', 'color', 'clarity','x','y','z','price'])].values
y_train=train_df["price"].values
y_test=test_df["price"].values;
````

## Step 2: Fit models and tune Hyperparamters using Training data

We will be using a random forest regressor. 
To train the model we will follow those steps:
* Create a dictionary of parameters
* Perform a gridsearchcv on the random forest model

````python
modreg= RandomForestRegressor(n_jobs=-1,random_state=2301)
'''
creating parameter list
'''
cv_params = {'n_estimators':[50,500,1000],'max_depth': [15,100,1000]}

'''
initializing Grid Search method
'''
gs = GridSearchCV(estimator=modreg,param_grid=cv_params,
                  scoring="neg_mean_squared_error",cv=10,return_train_score=False);
````

# Using Grid Search with Pipelines to search for best algorithms

* You can use pipelines along with Gridsearsh to search a space of algorithms and processing steps
* care needs to be taken to not create a very large search space

````python
'''
We fist define our pipeline
'''
pipe=Pipeline([('preprocessing', StandardScaler()), ('regressor', SVR())])

'''
We want to create a custom pre-pgrocessing step
'''
PCA_Pipe=Pipeline([('std_scaler', StandardScaler()),('pca',PCA(n_components=0.95,random_state=2301))])
'''

We now create the search space
'''

param_grid = [
    {'regressor': [SVR()], 'preprocessing': [StandardScaler(), PCA_Pipe],
     'regressor__epsilon': [0.1, 1, 10],
     'regressor__C': [0.1, 1, 10]},
    {'regressor': [RandomForestRegressor(random_state=2301)],
     'preprocessing': [None,PCA_Pipe], 'regressor__max_depth': [20, 50, 100],
    'regressor__n_estimators':[50,200,1000]}]

'''
We initialize the grid and the fit it
'''
grid = GridSearchCV(pipe, param_grid, cv=5,n_jobs=-1)
grid.fit(X_train, y_train);
````
